```python
from pyDOE import lhs
import matplotlib.pyplot as plt

# Generate 10 samples in 3D space
n_dimensions = 3
n_samples = 10
samples = lhs(n_dimensions, samples=n_samples)

# Plot the samples
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(samples[:, 0], samples[:, 1], samples[:, 2])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```

```python
# Generate random samples
import numpy as np
random_samples = np.random.rand(10, 2)

# Plot LHS samples and random samples for comparison
plt.figure(figsize=(10, 5))

# LHS samples
plt.subplot(1, 2, 1)
plt.scatter(samples[:, 0], samples[:, 1])
plt.title('Latin Hypercube Sampling')
plt.xlabel('X')
plt.ylabel('Y')

# Random samples
plt.subplot(1, 2, 2)
plt.scatter(random_samples[:, 0], random_samples[:, 1])
plt.title('Random Sampling')
plt.xlabel('X')
plt.ylabel('Y')

plt.tight_layout()
plt.show()
```
